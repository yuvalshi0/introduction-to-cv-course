```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parabole is ax**2 + bx + c
# this is why B = [a,b,c] , X = [x**2, x, 1] for each x in points and Y = [y1...,yn] for each y in points
points = np.asarray([(1, 10), (2, 5.49), (3, 0.89), (4, -0.14), (5, -1.07), (6, 0.84)])
x = points[:, 0]
y = points[:, 1]
Y = np.matrix(y).T  # one column with [y0.....,yn]
X = np.matrix(
    [[np.square(xi), xi, 1] for xi in x]
)  # create a row with the columns x**2, x**1, x**0 for each point
B = (X.T * X).I * (X.T * Y)  # or just np.polyfit(x, y, 2)
sample = np.linspace(0, 9)  # creaTe an array with the ranges 0 to 7 with 50 numbers
P = np.poly1d(np.asarray(B).flatten())  # create an equation from B
plt.title(f"Parabola [formula={P}]")
plt.plot(x, y, ".", sample, P(sample), "-")
plt.show()
```
